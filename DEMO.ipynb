```python
from pybel import BELGraph
from bio2bel_famplex.relations import enrich_graph
from pybel.dsl import protein, named_complex_abundance
from pybel_jupyter import to_jupyter
```

```python
graph = BELGraph(name="Demo", version="0.01", authors="Kristian Kolpeja")
graph.add_increases(
    protein("FPLX", "Actin"),
    protein("HGNC", "YFG"),
    "",
    "",
)
```

```python
enrich_graph(graph)
to_jupyter(graph)
```

```python
from pybel.struct import infer_child_relations
infer_child_relations(graph, protein("FPLX", "Actin"))
to_jupyter(graph)
```
